```python
# Q1. Explain GET and POST methods.
```

```python
"""
The GET and POST methods are HTTP request methods that are used to submit data from a web page to the server.

GET method is used to retrieve data from a server. 
When a user submits a GET request, the server retrieves the data and sends it back to the client. 
The data is sent in the URL string, which can be seen in the browser's address bar.

POST method is used to send data to the server to be processed. 
When a user submits a POST request, the data is sent in the request body, which is hidden from the user. 
This method is typically used when sensitive data needs to be sent, such as passwords or credit card information.
"""
```

```python
# Q3. Why is redirect() used in Flask?
```

```python
"""
The redirect() function is used in Flask to redirect the user to a different URL. 
It is typically used after the user submits a form, or after performing some action that requires a page reload. For example, 
if a user submits a form and the data is successfully processed, the redirect function can be used to redirect the user to a "success" page.
"""
```

```python
"""
In Flask, templates are used to generate dynamic HTML pages based on data provided by the application. 
Templates are typically written in HTML with placeholders for dynamic data that is supplied by the Flask application.

The render_template() function is used in Flask to render a template with data supplied by the application. 
The function takes the name of the template as its first argument, and any additional arguments represent the data to be supplied to the template.
"""
```
